# Question 1 — Categorical Scatter Plots

## Dataset

`tips`

## Problem Statement

Load the dataset into a DataFrame named `tips`.

### Requirements

Create the following two graphs:

1. Strip Plot
   - X-axis → `day`
   - Y-axis → `total_bill`
   - `hue='sex'`

2. Swarm Plot
   - X-axis → `day`
   - Y-axis → `tip`
   - `hue='time'`

Set figure size `(10,6)` for both graphs.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

# Question 7 — PairGrid

## Dataset

`penguins`

## Problem Statement

Load the dataset into a DataFrame named `penguins`.

### Requirements

Create a PairGrid.

- Upper Triangle → Scatter Plot
- Lower Triangle → KDE Plot
- Diagonal → Histogram
- `hue='species'`


```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
g = sns.PairGrid(data=penguins, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend()
g.fig.suptitle("Penguins Dataset: PairGrid Custom Subplots", y=1.02)
plt.show()
```

# Question 8 — FacetGrid

## Dataset

`tips`

## Problem Statement

Load the dataset into a DataFrame named `tips`.

### Requirements

Create a FacetGrid.

- Columns → `time`
- Rows → `sex`
- Plot Scatter Plot
  - X-axis → `total_bill`
  - Y-axis → `tip`

```python
tips = sns.load_dataset("tips")
tips
```

```python
g = sns.FacetGrid(data=tips, row="sex", col="time")
g.map(sns.scatterplot, "total_bill", "tip")
g.fig.suptitle("FacetGrid: Total Bill vs Tip by Sex and Time", y=1.02)
plt.show()
```

# Question 9 — Joint Plots

## Dataset

`penguins`

## Problem Statement

Load the dataset into a DataFrame named `penguins`.

### Requirements

Create the following graphs:

1. Joint Plot
   - X-axis → `bill_length_mm`
   - Y-axis → `bill_depth_mm`
   - `kind='hex'`

2. JointGrid
   - X-axis → `bill_length_mm`
   - Y-axis → `bill_depth_mm`
   - Scatter Plot in the center
   - Histogram on the top
   - Histogram on the right


```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
g1 = sns.jointplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="hex"
)
g1.fig.suptitle("Joint Plot: Bill Length vs Bill Depth", y=1.02)
plt.show()
```

```python
g2 = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g2.plot_joint(sns.scatterplot)
g2.plot_marginals(sns.histplot)
g2.fig.suptitle("JointGrid: Scatter Plot with Marginal Histograms", y=1.02)
plt.show()
```
