# Question 1 — Scatter Plot with Multiple Encodings

## Dataset

`tips`

## Problem Statement

Load the dataset into a DataFrame named `tips`.

### Requirements

Create a scatter plot using `sns.scatterplot()`.

- X-axis → `total_bill`
- Y-axis → `tip`
- Use:
  - `hue='sex'`
  - `style='time'`
  - `size='size'`
- Set figure size to `(10,6)`.
- Add:
  - Title
  - Grid


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
plt.title("Total Bill Vs Tips")
plt.grid()
plt.show()
```

# Question 4 — Facet Scatter Plot

## Dataset

`penguins`

## Problem Statement

Load the dataset into a DataFrame named `penguins`.

### Requirements

Create a figure-level scatter plot using `sns.relplot()`.

- X-axis → `bill_length_mm`
- Y-axis → `bill_depth_mm`
- Use:
  - `kind='scatter'`
  - `hue='species'`
  - `col='sex'`
  - `col_wrap=2`
- Add a title.

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
g = sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', col='sex', col_wrap=2, kind='scatter', height=5, aspect=1)
g.fig.suptitle("Bill Length vs Bill Depth by Species and Sex", fontsize=16)
g.fig.subplots_adjust(top=0.88)
plt.show()
```

# Question 5 — Heatmap

## Dataset

`flights`

## Problem Statement

Load the dataset into a DataFrame named `flights`.

### Requirements

1. Convert the dataset into a pivot table:
   - Index → `month`
   - Columns → `year`
   - Values → `passengers`
2. Create a heatmap using `sns.heatmap()`.
3. Display:
   - Cell values (`annot=True`)
   - Any colormap of your choice
4. Add:
   - Title
   - X-axis label
   - Y-axis label
5. Set figure size to `(12,6)`.

```python
flights =sns.load_dataset('flights')
flights
```

```python
flight_pivot = flights.pivot(index='month', columns='year', values='passengers')
flight_pivot
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flight_pivot, annot=True, cmap='YlGnBu', fmt="d")
plt.title("Monthly Airline Passengers Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
