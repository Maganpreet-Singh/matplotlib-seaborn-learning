# Question 1 — Basic Line Plot

## Dataset

`sharma-kohli.csv`

## Problem Statement

Load the dataset into a DataFrame named `players`.

### Requirements

1. Plot Rohit Sharma's season-wise runs.
2. Plot Virat Kohli's season-wise runs on the same graph.
3. Use different colors for both players.
4. Add:
   - Title
   - X-axis label
   - Y-axis label
   - Legend
   - Grid
5. Set figure size to `(10,5)`.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv("sharma-kohli.csv")
players
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    players["index"],
    players["RG Sharma"],
    color="blue",
    marker="o",
    linewidth=3,
    label="Rohit Sharma",
)
plt.plot(
    players["index"],
    players["V Kohli"],
    color="red",
    marker="s",
    linewidth=3,
    label="Virat Kohli",
)
plt.title("Rohit Sharma vs Virat Kohli - Season-wise Runs", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Runs Scored", fontsize=12)
plt.xticks(players["index"])
plt.legend()
plt.grid(True)
plt.show()
```

# Question 3 — Scatter Plot

## Dataset

## Problem Statement

Load the dataset into a DataFrame named `batter`.

### Requirements

   - Y-axis → Strike Rate
2. Add:

   - Grid

# Question 7 — Histogram of Batting Average

## Dataset

`batter.csv`

## Problem Statement

Load the dataset into a DataFrame named `batter`.

### Requirements

1. Plot a histogram of batting averages.
2. Use 10 bins.
3. Add:
   - Title
   - X-axis label
   - Y-axis label
   - Grid
4. Identify the average range having the maximum number of batters.

```python
batter == pd.read_csv('batter.csv')
batter
```

```python
plt.hist(
    batter['avg'],bins=10,color='teal',edgecolor='black'
)
plt.title("Distribution of Batting Averages", fontsize=14)
plt.xlabel("Batting Average", fontsize=12)
plt.ylabel("Number of Batters (Frequency)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
counts, bins = np.histogram(batter["avg"].dropna(), bins=10)
max_bin_index = np.argmax(counts)
lower_bound = bins[max_bin_index]
upper_bound = bins[max_bin_index + 1]

print(
    f"The average range with the maximum number of batters is approximately: {lower_bound:.2f} to {upper_bound:.2f} (Count: {counts[max_bin_index]})"
)
```

# Question 8 — Histogram of Strike Rate

## Dataset

`batter.csv`

## Problem Statement

Load the dataset into a DataFrame named `batter`.

### Requirements

1. Plot a histogram of strike rates.
2. Use 15 bins.
3. Add title, labels and grid.
4. Compare the spread of strike rate with batting average.

```python
batter = pd.read_csv('batter.csv')
batter
```

```python
plt.hist(
    batter["strike_rate"], bins=15, color="chocolate", edgecolor="black", alpha=0.8
)
plt.title("Distribution of Batting Strike Rates", fontsize=14)
plt.xlabel("Strike Rate", fontsize=12)
plt.ylabel("Number of Batters (Frequency)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
avg_mean, avg_std = batter["avg"].mean(), batter["avg"].std()
sr_mean, sr_std = batter["strike_rate"].mean(), batter["strike_rate"].std()
print("Spread Comparison:")
print(f"Batting Average - Mean: {avg_mean:.2f}, Std Dev: {avg_std:.2f}")
print(f"Strike Rate     - Mean: {sr_mean:.2f}, Std Dev: {sr_std:.2f}")
print(
    "\nObservation: Strike rates have a wider absolute range and spread (higher standard deviation) "
    "compared to batting averages, as they typically vary from under 100 to over 160, "
    "whereas batting averages are tightly bound between 10 and 50."
)
```

# Question 9 — Pie Chart

## Dataset

`gayle-175.csv`

## Problem Statement

Load the dataset into a DataFrame named `innings`.

### Requirements

1. Create a pie chart showing contribution of each batsman.
2. Display percentage values.
3. Start the chart at 90°.
4. Add title.
5. Determine:
   - Chris Gayle's contribution percentage.
   - Combined contribution of all remaining batsmen.

```python
innings = pd.read_csv('gayle-175.csv')
innings
```

```python
plt.pie(
    innings["batsman_runs"],
    labels=innings["batsman"],
    autopct="%1.1f%%",
    startangle=90,
    colors=["lightcoral", "skyblue", "lightgreen", "gold","teal"],
)
plt.title("Batsman Contribution - Chris Gayle's 175* Innings", fontsize=14)
plt.show()
total_runs = innings["batsman_runs"].sum()
gayle_runs = innings.loc[innings["batsman"] == "CH Gayle", "batsman_runs"].values[0]
remaining_runs = total_runs - gayle_runs

gayle_pct = (gayle_runs / total_runs) * 100
remaining_pct = (remaining_runs / total_runs) * 100

print(f"Chris Gayle's contribution percentage: {gayle_pct:.1f}%")
print(
    f"Combined contribution of all remaining batsmen: {remaining_pct:.1f}%"
)
```
